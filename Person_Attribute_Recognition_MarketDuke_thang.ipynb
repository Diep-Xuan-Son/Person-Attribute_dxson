```python
%cd /content/drive/MyDrive/Person_Attribute
import json
import os, glob
import sys
with open('./doc/person_final_label.json', 'r') as labelfile:
    label = json.load(labelfile)["label"]
label.sort()
print(label)
for root, dirs, files in os.walk("/content/drive/MyDrive/Person_Attribute/dataset/dataset/jsonfile"):
  for file in files:
    if file.endswith(".json"):
      name_json = file.split(".")[0]
      with open(os.path.join(root, file), "r") as filejson:
        globals()[f"data_{name_json}"] = json.load(filejson)
data = dict()
data['train_att'] = dict()
data['test_att'] = dict()
inter = 0
data_img_train = []
data_img_test = []
dir_ids = []
for root, dirs, files in os.walk('/content/drive/MyDrive/Person_Attribute/dataset/dataset'):
    for file in files:
        if file.endswith('.jpg'):
            inter += 1
            arr = []
            print("root.split('/') la:", root.split('/'))  
            # root.split('/') la: ['', 'content', 'drive', 'MyDrive', 'Person_Attribute', 'dataset', 'dataset', '13_04_2022', '13_04_2022', 'ids_876']
            print("root.split('/'){} la:", root.split('/')[8:])
            name_dataset, ids_folder = root.split('/')[8:]
            ids = f'{name_dataset}_'+ids_folder
            img = os.path.join(root, file)
            dir_ids.append(ids)
            cam = 1
            name = file.replace('.jpg', '')
            for i in label:
                arr.append(globals()[f'data_{name_dataset}'][ids_folder][i])
            if inter %10 == 0:
                data_img_test.append([img, dir_ids.index(ids), ids, cam, name])
                data['test_att'][ids] = arr
            data_img_train.append([img, dir_ids.index(ids), ids, cam, name])
            data['train_att'][ids] = arr
data['test_img'] = {'data':data_img_test, 'ids':dir_ids}
data['train_img'] = {'data':data_img_train, 'ids':dir_ids}
data['label'] = label
for keys, value in data['train_att'].items():
    print(keys, value)
    break
with open('./doc/data.json', 'w') as f:
    json.dump(data, f, indent = 4)
```

```python
import xlrd
import json
```

```python
loc = ("../dataset/18_04_2022/18_04_2022.xlsx")
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0)
# For row 0 and column 0
label = sheet.row_values(0)[1:]
result = dict()
for i in range(1, sheet.nrows):
  data = sheet.row_values(i)
  ids = data[0]
  result[ids] = dict()
  for inter, i in enumerate(data[1:]):
    if i == 1:
      result[ids][label[inter]] = 1
    else:
      result[ids][label[inter]] = 0
with open("../dataset/jsonfile/18_04_2022.json", "w") as f:
	json.dump(result,f, indent = 4)
 
with open("./doc/person_final_label.json", "w") as f:
  json.dump({"label": label}, f, indent = 4)
```
